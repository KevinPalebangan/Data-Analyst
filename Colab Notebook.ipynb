# **Visualization & Explanatory Analysis**

# Pertanyaan 1 : Produk apa yang dominan terjual?

```python
sum_order_items_dataset_df = all_data.groupby("product_category_name_english")["product_id"].count().reset_index()
sum_order_items_dataset_df = sum_order_items_dataset_df.rename(columns={"product_id": "products"})
sum_order_items_dataset_df = sum_order_items_dataset_df.sort_values(by="products", ascending=False)
sum_order_items_dataset_df = sum_order_items_dataset_df.head(10)
sum_order_items_df.head()
```

*Pada bagian ini menampilkan grafik antara produk yang dominan dan paling minim terjual, agar kita bisa mengetahui secara rinci manakah produk yang paling dominan terujal*

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(24, 6))

colors = ["#068DA9", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]

sns.barplot(x="products", y="product_category_name_english", data=sum_order_items_dataset_df.head(5), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("Produk Yang Dominan Terjual:", loc="center", fontsize=20)
ax[0].tick_params(axis ='y', labelsize=15)

sns.barplot(x="products", y="product_category_name_english", data=sum_order_items_dataset_df.sort_values(by="products", ascending=True).head(5), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].set_title("Produk Yang Minim Terjual", loc="center", fontsize=20)
ax[1].tick_params(axis='y', labelsize=15)

plt.suptitle("Produk Dominan Terjual", fontsize=30)
plt.show()
```

*Dari data diatas menampilkan grafik antara produk yang dominan dan paling minim terjual, agar kita bisa mengetahui secara rinci manakah produk yang paling dominan terujal, yang dimana produk yang paling dominan terjual ialah bed bath table*

# Pertanyaan 2 : Apakah customer puas dengan layanan e-commerce kami?

```python
review_scores = all_data['review_score'].value_counts().sort_values(ascending=False)

most_common_score = review_scores.idxmax()

sns.set(style="darkgrid")

plt.figure(figsize=(8, 4))
sns.barplot(x=review_scores.index,
            y=review_scores.values,
            order=review_scores.index,
            palette=["#068DA9" if score == most_common_score else "#D3D3D3" for score in review_scores.index]
            )

plt.title("Tingkat Kepuasan Layanan E-Commerce", fontsize=15)
plt.xlabel("Rate")
plt.ylabel("Jumlah")
plt.xticks(fontsize=12)
plt.show()
```

*Dari grafik diatas kita dapat mengetahui bahwa sebagian besar customer memberikan rating 5*
